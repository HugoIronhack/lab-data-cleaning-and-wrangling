### We will start with removing outliers. So far, we have discussed different methods to remove outliers. Use the one you feel more comfortable with, define a function for that. 
### Use the function to remove the outliers and apply it to the dataframe.

```python
def plot_boxplots(df):  # boxplot makes it easy to see outliers
    for col in df.columns:
        plt.figure(figsize=(8,6))
        sns.boxplot(x=df[col])
        plt.title(col)
        plt.show()
```

```python
plot_boxplots(numerical_df)
```

```python
# many outliers among Customer lifetime value, Monthly Premium Auto and Total Claim Amount
```

```python
customer_df_2 = customer_df.copy() # i make a copy first and then remove outliers
```

```python

def remove_outliers(data):
    # Calculate the first and third quartiles
    q1, q3 = np.percentile(data, [25, 75])
    
    # Calculate the interquartile range
    iqr = q3 - q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Remove outliers from the data
    cleaned_data = [x for x in data if lower_bound <= x <= upper_bound]
    
    return cleaned_data
```

```python
cleaned_df = numerical_df.apply(remove_outliers, axis=0)
```

```python
from sklearn.preprocessing import StandardScaler

# Extract the continuous variables from the DataFrame
continuous_vars = customer_df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the continuous variables
scaler.fit(customer_df[continuous_vars])

# Transform the continuous variables using the scaler
customer_df[continuous_vars] = scaler.transform(customer_df[continuous_vars])
```

```python
customer_df[continuous_vars]
```

### Encode the categorical variables

```python
# Extract the categorical variables from the DataFrame
categorical_vars = customer_df.select_dtypes(include=['object']).columns.tolist()

# Encode the categorical variables using get_dummies
encoded_df = pd.get_dummies(customer_df, columns=categorical_vars, drop_first=True)
```
